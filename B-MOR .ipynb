```python
import pandas as pd
import matplotlib.pyplot as plt

langs= ['RidgeCV', '2', '4', '6', '8']

threads2 = [3740, 2304 ,1395, 1090 ,913]
threads4 = [2380, 1355, 877, 664, 574]
threads8 = [1643, 891, 572, 435, 384]
threads16 = [1112, 648, 431, 362, 303]
threads32 = [1012, 547, 384, 319, 287]


df = pd.DataFrame({
    '#MKL threads=2': threads2,
    '#MKL threads=4': threads4,
    '#MKL threads=8': threads8,
    '#MKL threads=16':threads16,
    '#MKL threads=32':threads32
}, index=langs)

error2 = [80, 80, 80, 80, 80]

ax = df.plot.bar(figsize=(22, 12), fontsize=40, yerr=error2, rot=0, width=0.9, edgecolor='black')

# Specify columns to have '+' patterns
plus_cols = ['#MKL threads=2']
for col in plus_cols:
    for i, value in enumerate(df[col]):
        ax.patches[i * len(df.columns) + df.columns.get_loc(col)].set_hatch('+')

# Extend legend with 10 patterns and different background colors
legend_labels = [
   '#MKL threads=2', '#Dask_MKL threads=2',  '#MKL threads=4',  '#Dask_MKL threads=4',              
                    '#MKL threads=8' ,  '#Dask_MKL threads=8',
                   '#MKL threads=16',    '#Dask_MKL threads=16',  
                    '#MKL threads=32' ,    '#Dask_ MKL threads=32' ]

legend_handles = [plt.Rectangle((0, 0), 1, 1, fc=f'C{i//2}', hatch='+' if i % 2 == 1 else '') for i in range(10)]

ax.legend(legend_handles, legend_labels, loc='upper right', bbox_to_anchor=(1, 1), fontsize=28)

plt.xlabel('Number of Dask_Workers', fontsize=45)
plt.ylabel('Training time (Seconds)', fontsize=45)


plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sub1
threads_2 = [3740, 2304, 1395, 1090, 913]
threads_4 = [2380, 1355, 877, 664, 574]
threads_8 = [1643, 891, 572, 435, 384]
threads_16 = [1112, 648, 431, 362, 303]
threads_32 = [1012, 547, 384, 319, 287]

# Use the execution time with 1 worker and 2 threads as the reference execution time
reference_execution_time = threads_2[0]


data = np.array([threads_2, threads_4, threads_8, threads_16, threads_32])


speedup_data = np.zeros_like(data, dtype=float)

for i in range(len(data)):  
    for j in range(len(data[i])):  
        speedup_data[i][j] = reference_execution_time / data[i][j]


plt.rcParams.update({'font.size': 14})

fig, ax = plt.subplots(figsize=(10, 6))

im = ax.imshow(speedup_data, cmap='viridis', interpolation='nearest', origin='lower')


ax.set_xlabel('Number of Threads',fontsize='25')
ax.set_ylabel('Number of Workers',fontsize='25')
ax.set_title('Speed-Up Matrix',fontsize='25')


ax.set_xticks(np.arange(len(threads_2)))
ax.set_yticks(np.arange(len([1, 2, 4, 6, 8])))
ax.set_xticklabels([2, 4, 8, 16, 32],fontsize='25')
ax.set_yticklabels([1, 2, 4, 6, 8],fontsize='25')


cbar = plt.colorbar(im, ax=ax)
max_value = 16
im.set_clim(vmin=0, vmax=max_value)

cbar.ax.tick_params(labelsize=26)
#cbar.remove()

plt.show()
```
