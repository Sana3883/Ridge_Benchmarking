```python
import pandas as pd
import matplotlib.pyplot as plt

langs= ['2', '4', '6', '8']

threads2 = [6261, 3270, 2328, 1781]
threads4 = [4418, 2289, 1605, 1201]
threads8 = [3350, 1790, 1212, 925]
threads16 = [3122, 1578, 1112, 790]
threads32 = [2835, 1451, 985, 720]


df = pd.DataFrame({ '#Dask_MKL threads=2': threads2,
                    '#Dask_MKL threads=4': threads4,                    
                    '#Dask_MKL threads=8' : threads8,
                   '#Dask_MKL threads=16' : threads16,  
                    '#Dask_MKL threads=32' : threads32
                  }, index=langs)


error2=[50,50,50,50]
    

df.plot.bar(figsize=(22,12), fontsize=40, yerr=error2, rot=0,  width=0.9);

plt.xlabel('Number of Dask_Workers', fontsize=40)
plt.ylabel('Training time (Seconds)', fontsize=40)
plt.legend(fontsize=30)
```
