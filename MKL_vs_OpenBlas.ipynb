```python
import matplotlib.pyplot as plt

# Data for the execution time of Ridge for different numbers of threads
subjects = ["Subject6", "Subject5", "Subject4", "Subject3", "Subject2", "Subject1"]
num_processors = [4, 8, 16, 32]

mkl_execution_times = [
    [1761, 1127, 792, 647],
    [1766, 1113, 790, 622],
    [1769, 1149, 795, 618],
    [1749, 1189, 765, 628],
    [1734, 1104, 782, 651],
    [1723, 1136, 762, 642]
]

openblas_execution_times = [
    [2252, 1592, 1357, 1212],
    [2229, 1614, 1260, 1162],
    [2295, 1622, 1342, 1205],
    [2265, 1692, 1382, 1285],
    [2288, 1588, 1301, 1176],
    [2230, 1600, 1268, 1165]
]


# Calculate speed-up for each subject and library
mkl_speedup = [[mkl_execution_times[i][0] / time for time in mkl_execution_times[i]] for i in range(len(subjects))]
openblas_speedup = [[openblas_execution_times[i][0] / time for time in openblas_execution_times[i]] for i in range(len(subjects))]

# Define four distinct colors
colors = ['b', 'g', 'r', 'c', 'm', 'y']

# Plotting
plt.figure(figsize=(12, 8))

for i in range(len(subjects)):
    plt.plot(num_processors, mkl_speedup[i], label=f'MKL - {subjects[i]}', marker='o', color=colors[i])
    plt.plot(num_processors, openblas_speedup[i], label=f'OpenBLAS - {subjects[i]}', marker='o', color=colors[i], linestyle='--')

plt.xlabel('Number of Processors', fontsize=20)
plt.ylabel('Parallel Efficiency (Speed-Up / Number of Processors)', fontsize=20)
#plt.title('ROI-wise', fontsize=20)
plt.legend(fontsize=12)
plt.grid(True)

# Increase the font size of numbers on the axes
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)

plt.show()
```
